```python
import pandas as pd 
import os
from pandarallel import pandarallel
pandarallel.initialize()

DATA = os.getcwd()

train_raw = pd.read_csv ( DATA + "/train.csv")
test_raw = pd.read_csv ( DATA + "/test.csv")


for df in [train_raw, test_raw]:
    df['group']     = df['PassengerId'].parallel_apply(lambda x: x.split('_')[0])
    df['member']     = df['PassengerId'].parallel_apply(lambda x: x.split('_')[1])
    cabin_split = df['Cabin'].parallel_apply(lambda x: str(x).split('/'))
    df['deck'] = cabin_split[0][0]
    df['deck_number'] = cabin_split[0][1]
    df['deck_side'] = cabin_split[0][2]
    
    df.drop(['PassengerId','Cabin'] , inplace=True , axis= 1)

    
display ( train_raw )
```
